```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

### Función Análisis Exploratorio de Datos

```python
# Funcion para un Análisis Exploratorio inicial de los datos
def exploracion_inicial(df, tipo=None):
    print("¿Cuántas filas y columnas hay en el conjunto de datos?")
    num_filas, num_columnas = df.shape
    print(f"\tHay {num_filas:,} filas y {num_columnas:,} columnas.")
    print('#' * 90)

    if tipo == 'simple':
        print("¿Cuáles son las primeras dos filas del conjunto de datos?")
        display(df.head(2))
    else:
        print("¿Cuáles son las primeras cinco filas del conjunto de datos?")
        display(df.head())

        print("¿Cuáles son las últimas cinco filas del conjunto de datos?")
        display(df.tail())

        print("¿Cómo puedes obtener una muestra aleatoria de filas del conjunto de datos?")
        display(df.sample(n=5))

        print("¿Cuáles son las columnas del conjunto de datos?")
        print("\n".join(f"\t- {col}" for col in df.columns))

        print("¿Cuál es el tipo de datos de cada columna?")
        print(df.dtypes)

        print("¿Cuántas columnas hay de cada tipo de datos?")
        print(df.dtypes.value_counts())

        print("¿Cómo podríamos obtener información más completa sobre la estructura y el contenido del DataFrame?")
        df.info()

        print("¿Cuántos valores únicos tiene cada columna?")
        print(df.nunique())

        print("¿Cuáles son las estadísticas descriptivas básicas de todas las columnas?")
        display(df.describe(include='all').fillna(''))

        print("¿Hay valores nulos en el conjunto de datos?")
        print(df.isnull().sum())

        print("¿Cuál es la proporción de valores nulos en cada columna?")
        print((df.isnull().sum() / len(df) * 100).round(2))

    print('#' * 90)
```

#### Comprobación de Nulos

En este Dataset existen Nulos en las siguientes variables: review_comment_title (87656 - 88.34%) y review_comment_message (58247 - 58.70%). Imputamos "No Title" y "No Comment" respectivamente

```python
df_reviews.fillna({'review_comment_title': 'No Title', 'review_comment_message': 'No Comment'}, inplace=True)
```

```python
df_reviews.isnull().sum()
```

#### Formateamos las variables de fecha a datetime

```python
df_reviews['review_creation_date'] = pd.to_datetime(df_reviews['review_creation_date']).dt.strftime('%Y-%m-%d')
df_reviews['review_answer_timestamp'] = pd.to_datetime(df_reviews['review_answer_timestamp']).dt.strftime('%Y-%m-%d')
```

### Representaciones Gráficas

#### Distribución Reviews

```python
plt.figure(figsize=(8, 5))
sns.histplot(df_reviews["review_score"], bins=6, discrete=True, kde=False, palette='viridiris')
plt.title("Distribución de Review Scores")
plt.xlabel("Score")
plt.ylabel("Frecuencia")
plt.tight_layout()
plt.show()
```

### Promedio Review_Score por Mes

```python
# Extraemos Año y Mes
df_reviews["year"] = df_reviews["review_creation_date"].dt.year
df_reviews["month"] = df_reviews["review_creation_date"].dt.month

# Agrupamos por Mes y Año
mean_scores_by_month = df_reviews.groupby(["year", "month"])["review_score"].mean().reset_index()

# Creamos una columna combinada de año y mes para el eje X
mean_scores_by_month["year_month"] = pd.to_datetime(mean_scores_by_month[["year", "month"]].assign(day=1))

plt.figure(figsize=(12, 6))
plt.plot(mean_scores_by_month["year_month"], mean_scores_by_month["review_score"], marker='o')
plt.title("Media de Review Score por Mes")
plt.xlabel("Fecha")
plt.ylabel("Review Score Promedio")
plt.grid(True)
plt.tight_layout()
plt.show()
```
